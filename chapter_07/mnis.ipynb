```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = "../data/mnist/mnist_experiments.csv"
```

```python
df = pd.read_csv(data)
```

```python
df
```

```python
fig = sns.catplot(data=df, kind="bar", x="model", y="score", row="run", color="blue", alpha=0.7)

fig.fig.set_size_inches((12,15))
fig.axes[-1][-1].set_xticklabels(labels=df.model.unique(),rotation=90);

for ax in fig.axes:
    ax[0].set_ylim(0.5, 1)
```

```python
for name, grp in df.groupby("model"):
    fig = sns.catplot(data=grp, kind="bar", x="run", y="score", alpha=0.7)
    fig.ax.set_title(grp.model.iloc[0])
    fig.ax.set_ylim(0.5, 1)
```
